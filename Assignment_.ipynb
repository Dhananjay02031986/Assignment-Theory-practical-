```python

#Theory  questions:
#1-what is the difference between a funtion and a method in pyhton ?
. Function:
A function is a block of reusable code that is designed to perform a specific task.
It can be defined anywhere and isn't tied to a particular object.
You can call a function by its name directly.

def greet(name):
    return f"Hello, {name}!"

Method:
A method is a function that is associated with an object (or class).
It’s essentially a function that belongs to an instance of a class and operates on the data that is contained in that instance.
Methods are called on an object, and they always have at least one argument, usually self, which refers to the object itself
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

# Creating an instance of the class
person = Person("Alice")

# Calling the method
print(person.greet())  # "Hello, Alice!"


#2-explain the concept of function arguments and parameters in python?
In Python, function arguments and parameters are key concepts that help pass data into functions so they can operate on it.
They are used interchangeably in casual conversation, but they have specific meanings when discussing functions.
Parameters:
Parameters are the variables that are listed inside the parentheses in a function definition.
They are essentially "placeholders" for the values that will be passed to the function when it is called
def greet(name, age):  # 'name' and 'age' are parameters
    return f"Hello, {name}! You are {age} years old."
Arguments:
Arguments are the actual values that are passed into the function when you call it
They correspond to the parameters in the function definition.
print(greet("Alice", 30))  # "Alice" and 30 are arguments
Here, "Alice" and 30 are arguments that match the parameters name and age in the function.




#3-what are the different ways to define and call a function in pyhton?
1-Standard Function Definition:
The most common way to define a function is with the def keyword, followed by the function name and parameters inside parentheses.
def greet(name):
    return f"Hello, {name}!"
    print(greet("Alice"))


2. Functions with Default Parameters:
You can define default values for parameters in the function signature. If no argument is provided, the default value is used.
def greet(name="Guest"):
    return f"Hello, {name}!"
print(greet())         # "Hello, Guest!"
print(greet("Alice"))  # "Hello, Alice!"

3. Functions with Variable-Length Arguments
*args: Accepts a variable number of positional arguments.
**kwargs: Accepts a variable number of keyword arguments.
def summarize(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

    summarize(1, 2, 3, name="Alice", age=30)
# Output:
# Positional arguments: (1, 2, 3)
# Keyword arguments: {'name': 'Alice', 'age': 30}
4. Lambda Functions (Anonymous Functions):
Lambda functions are small, unnamed functions that are defined using the lambda keyword.
 These are typically used for short-lived, one-line functions.
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8



#4-what is purpose of the 'return'statement in python function?
The return statement in Python serves a key purpose: it terminates the function and sends a result back to the caller.
 This allows functions to output values that can be used elsewhere in your program.
 Without a return statement, a function will not return any value by default, and it implicitly returns None.

Purpose of the return statement:
1-Returning a Value to the Caller:The main purpose of the return statement is to return a result from the function to where it was called.
This makes functions more useful because they can produce values that can be assigned to variables, printed, or further manipulated.

def add(a, b):
    return a + b

result = add(5, 3)  # 'result' gets the returned value
print(result)        # Output: 8

2.	Exiting the Function Early
o	A return statement can also be used to exit a function early.
This is useful if a condition is met, and no further code should be executed in the function
def divide(a, b):
    if b == 0:
        return "Error: Division by zero"  # Exits function early
    return a / b

print(divide(10, 2))  # Output: 5.0
print(divide(10, 0))  # Output: Error: Division by zero

3.	Returning Multiple Values:
o	Python allows you to return multiple values at once, which are packed into a tuple.
 This is handy when you need to return more than one result from a function.
 def get_name_and_age():
    return "Alice", 30

name, age = get_name_and_age()
print(name)  # Output: Alice
print(age)   # Output: 30
the function get_name_and_age() returns two values ("Alice" and 30). Python automatically packs them into a tuple (name, age).
4.	Returning None Implicitly:
	If a function doesn't have a return statement, or if the return statement has no expression (i.e., return alone), the function will implicitly return None.
  def greet(name):
    print(f"Hello, {name}!")

result = greet("Alice")
print(result)  # Output: None
Since the greet function doesn’t have a return statement, it implicitly returns None.
5.	Returning a Value for Further Use:
The value returned can be assigned to a variable or used in further expressions, allowing for complex processing in one line.
Example:
def multiply(a, b):
    return a * b

result = multiply(4, 5) + 10  # Uses the returned value in further calculations
print(result)  # Output: 30
Here, multiply(4, 5) returns 20, and the result is then used in the expression 20 + 10.


#5-what are iterators in python and how do they differ from iterables?
1. Iterable:
An iterable is any Python object capable of returning its members one at a time, allowing it to be iterated over (looped through) in a for-loop, or used with functions like iter() and next().
•	Definition: An iterable is an object that implements the __iter__() method or the __getitem__() method (for older Python versions).
•	Examples of iterables: lists, tuples, strings, dictionaries, sets, etc.
•	What makes it iterable: The ability to return an iterator when passed to the iter() function.
my_list = [1, 2, 3]

# You can iterate over the list (an iterable) using a for-loop
for item in my_list:
    print(item)

2. Iterator:
An iterator is an object representing a stream of data; it produces the next value when the next() function is called on it.
 It is the actual object that keeps track of the iteration process, meaning it "knows" the next value to return.
 -	Definition: An iterator is an object that implements both the __iter__() and __next__() methods.
-	The __iter__() method returns the iterator object itself (this is useful for chaining iterators).
-	The __next__() method returns the next value in the sequence. When there are no more values left, it raises the StopIteration exception.

my_list = [1, 2, 3]

# Get an iterator from the iterable (list)
my_iter = iter(my_list)

# Now we can get each item one by one using the `next()` function
print(next(my_iter))  # Output: 1
print(next(my_iter))  # Output: 2
print(next(my_iter))  # Output: 3
print(next(my_iter))  # Raises StopIteration error
In this case:
•	my_list is an iterable.
•	my_iter is an iterator.


#6-explain the concept of generators in python and how they are defined?
A generator in Python is a type of iterable object that allows you to iterate over a sequence of values. Unlike lists or other data structures, generators generate values on the fly rather than storing them all in memory at once.
This is why they are referred to as lazy iterables: they produce values only when requested, making them more memory-efficient for large datasets or infinite sequences.

Generators are defined using either:

1-A function with yield.

2-A generator expression (similar to list comprehensions but with parentheses).

def simple_generator():
    yield 1
    yield 2
    yield 3

# Create the generator object
gen = simple_generator()

# Iterate over the generator
for value in gen:
    print(value)
Output:

1
2
3


#7-what are the advatage of using generators over regular functions?
Generators offer several key benefits over regular functions, especially
when working with large data sets or streams of data. Here are the primary advantages
a-Generators produce values one at a time and only when requested, meaning they do not store the entire sequence in memory.
This can be a huge advantage when working with large datasets.
b-Because generators only compute values on the fly and do not evaluate everything at once, they can start producing results immediately.
In contrast, functions that create large lists or data structures must wait until the entire structure is created, which can delay the process.

#8-what is lambda funtion in python and when it is typically used?
A lambda function is a small, anonymous function in Python that is defined using the lambda keyword.
When you need a quick, simple function that is not reused elsewhere.
Lambda functions are useful in places where defining a full function might seem overkill.

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))

print(squared_numbers)  # Output: [1, 4, 9, 16, 25]


#9-Explain the purpose and usage of the "map()"function in python
The map() function in Python is a built-in function used to apply a given function to all items in an iterable (such as a list, tuple, etc.)
and return a map object (an iterator) that yields the results
It is commonly used when you need to apply a function to every element of an iterable (like a list, tuple, etc.).

words = ['apple', 'banana', 'cherry']
word_lengths = list(map(len, words))

print(word_lengths)  # Output: [5, 6, 6]
print(list(uppercase_words))  # Output: ['APPLE', 'BANANA', 'CHERRY']


#10-what is difference between  'map()','reduced()' and 'filter()' function in python?
The map() function applies a given function to each item of an iterable (like a list, tuple, etc.)
and returns a new iterable (a map object) that contains the results.
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16, 25]


The filter() function is used to filter out elements from an iterable based on a given condition (function).
It only returns the items for which the function returns True.
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4, 6]

The reduce() function is used to reduce a list (or iterable) to a single cumulative result by applying a binary function (a function that takes two arguments) cumulatively to the items of the iterable.
numbers = [1, 2, 3, 4, 5]
sum_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_numbers)  # Output: 15 (1+2+3+4+5)

#11-using pen and paper write the internal mechanisum for sum operation using reduce funtion on this given list
list:[47,11,42,13];
reduce function from functools module:
The reduce() function takes a binary function and a sequence (like a list) and applies the binary function cumulatively to the items in the sequence. In this case, we will use it to perform a sum operation on a list.

Internal Mechanism of reduce for Sum Operation
Input: A list of numbers: [47, 11, 42, 13]

Function: The function we need to reduce the list with is a binary addition function that takes two arguments and returns their sum. We can define this as lambda x, y: x + y.

Execution: The reduce function applies the binary function step by step as follows:

Start with the first two elements of the list: 47 and 11.

Apply the function to them: 47 + 11 = 58.

Next, take the result (58) and apply the function to it with the next element (42): 58 + 42 = 100.

Finally, apply the function to the result (100) and the last element (13): 100 + 13 = 113.

Output: The final result after all reductions is 113.

Code Implementation
python
Copy
Edit
from functools import reduce

# Given list
lst = [47, 11, 42, 13]

# Using reduce to sum the elements
sum_result = reduce(lambda x, y: x + y, lst)

# Output the result
print("Sum of list:", sum_result)
Internal Flow of reduce:
First step: reduce(lambda x, y: x + y, [47, 11, 42, 13])

Initial list: [47, 11, 42, 13]

Apply lambda(47, 11) → 47 + 11 = 58

Partial result: 58

Second step: lambda(58, 42) → 58 + 42 = 100

Partial result: 100

Third step: lambda(100, 13) → 100 + 13 = 113

Final result: 113

So, the result is 113, which is the sum of the elements in the list.
```
